```python
from PIL import Image
from matplotlib.pylab import *

image = array(Image.open('images/profile.jpg').convert('L'))
```

```python
image
```

```python
image_negative = 255 - image
gray()
imshow(image_negative)
show()
```

```python
image_clamped = (100.0/255.0) * image + 100
imshow(image_clamped)
show()
```

```python
image_squared = 255.0 * (image / 255.0) ** 2
imshow(image_squared)
show()
```
