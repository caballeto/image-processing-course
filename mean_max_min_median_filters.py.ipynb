```python
import scipy.ndimage
import numpy as np
from PIL import Image

image = Image.open('images/lena_noisy.png').convert('L')

# mean filter
kernel = np.ones((5, 5)) / 25
result = scipy.ndimage.filters.convolve(image, kernel)
result = Image.fromarray(result)
result
```

```python
image = Image.open('images/lena_noisy.png')

# min filter
result = scipy.ndimage.filters.minimum_filter(image, size=5, footprint=None, output=None, mode='reflect', cval=0.0, origin=0)
result = Image.fromarray(result)
result
```

```python
image = Image.open('images/lena_noisy.png')

# max filter
result = scipy.ndimage.filters.maximum_filter(image, size=5, footprint=None, output=None, mode='reflect', cval=0.0, origin=0)
result = Image.fromarray(result)
result
```

```python
image = Image.open('images/lena_noisy.png')

# median filter
result = scipy.ndimage.filters.median_filter(image, size=5, footprint=None, output=None, mode='reflect', cval=0.0, origin=0)
result = Image.fromarray(result)
result
```
