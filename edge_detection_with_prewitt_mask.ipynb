```python
import cv2
import numpy as np

from skimage import img_as_int, img_as_ubyte
from scipy import ndimage
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('images/moon.png', 0)
image = img_as_int(image)
image
```

```python
prewitt_vertical = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype='float64')
result = img_as_ubyte(ndimage.convolve(image, prewitt_vertical, mode='constant', cval=0.0))
plt.imshow(result, cmap='gray')
plt.show()
```
